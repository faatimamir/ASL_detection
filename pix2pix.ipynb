# Install

```python
import PIL
import torch
from diffusers import StableDiffusionInstructPix2PixPipeline, EulerAncestralDiscreteScheduler

# Initialize the model
model_id = "timbrooks/instruct-pix2pix"
pipe = StableDiffusionInstructPix2PixPipeline.from_pretrained(
    model_id, torch_dtype=torch.float16, safety_checker=None
)
pipe.to("cuda")
pipe.scheduler = EulerAncestralDiscreteScheduler.from_config(pipe.scheduler.config)

# Load the image using PIL
image_path = "/content/black_10.jpg"  # Make sure the path is correct
image = PIL.Image.open(image_path).convert("RGB")  # Ensure the image is in RGB format

# Provide the prompt
prompt = "just add design on t-shirt"

# Run the pipeline
images = pipe(prompt=prompt, image=image, num_inference_steps=10, image_guidance_scale=2).images

# Display or save the generated image
images[0].show()  # Use .show() to display or save using images[0].save("output.jpg")
```
